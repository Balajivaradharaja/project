```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, precision_recall_curve
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, LSTM, Dense, RepeatVector, TimeDistributed, Multiply, Lambda
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import tensorflow as tf
import joblib
import warnings

warnings.filterwarnings('ignore')


def load_and_preprocess_data(file_path):
    # Load data
    df = pd.read_csv(file_path)
    
   
    api_logs = df[df['Log_Type'] == 'API'].dropna(subset=["Status_Code", "Processing_Time_ms"])
    
   
    if 'Timestamp' in api_logs.columns:
        api_logs['Timestamp'] = pd.to_datetime(api_logs['Timestamp'])
        api_logs.sort_values(by='Timestamp', inplace=True)
    
    
    api_logs['Request_Rate'] = api_logs['Timestamp'].diff().dt.total_seconds().fillna(0).apply(lambda x: 1 if x == 0 else 1/x)
    api_logs['Hour_Of_Day'] = api_logs['Timestamp'].dt.hour
    api_logs['Day_Of_Week'] = api_logs['Timestamp'].dt.dayofweek
    
   
    features = ["Status_Code", "Processing_Time_ms", "Request_Rate", "Hour_Of_Day", "Day_Of_Week"]
    
    
    scaler = RobustScaler()
    data_scaled = scaler.fit_transform(api_logs[features])
    joblib.dump(scaler, 'scaler.joblib')
    
    return data_scaled, api_logs


def create_sequences(data, window_size, stride=1):
    sequences = []
    for i in range(0, len(data) - window_size + 1, stride):
        seq = data[i:i+window_size]
        sequences.append(seq)
    return np.array(sequences)


def build_lstm_autoencoder(input_shape):
    inputs = Input(shape=input_shape)
    
    # Encoder
    x = LSTM(128, activation='relu', return_sequences=True)(inputs)
    x = LSTM(64, activation='relu', return_sequences=True)(x)
    x = LSTM(32, activation='relu', return_sequences=False)(x)
    
   
    attention_input = RepeatVector(input_shape[0])(x)
    attention = LSTM(32, activation='relu', return_sequences=True)(attention_input)
    attention = Dense(1, activation='tanh')(attention)
    attention = tf.keras.activations.softmax(attention, axis=1)
    
    
    context = Multiply()([attention_input, attention])
    context = Lambda(lambda x: tf.reduce_sum(x, axis=1))(context)
    
   
    repeated_context = RepeatVector(input_shape[0])(context)
    
 
    x = LSTM(32, activation='relu', return_sequences=True)(repeated_context)
    x = LSTM(64, activation='relu', return_sequences=True)(x)
    x = LSTM(128, activation='relu', return_sequences=True)(x)
    outputs = TimeDistributed(Dense(input_shape[1]))(x)
    
    model = Model(inputs, outputs)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    return model


def train_model(model, X_train, X_val):
    es = EarlyStopping(monitor='val_loss', patience=10, mode='min', restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
    
    history = model.fit(
        X_train, X_train,
        epochs=100,
        batch_size=64,
        validation_data=(X_val, X_val),
        callbacks=[es, reduce_lr]
    )
    return history


def evaluate_model(model, X_test, threshold_percentile=95):
    X_test_pred = model.predict(X_test)
    test_mse = np.mean(np.power(X_test.reshape(X_test.shape[0], -1) - 
                                X_test_pred.reshape(X_test_pred.shape[0], -1), 2), axis=1)
    
    threshold = np.percentile(test_mse, threshold_percentile)
    anomaly_flags = test_mse > threshold
    return test_mse, threshold, anomaly_flags

def generate_anomaly_report(test_mse, threshold, anomaly_flags, api_logs, window_size):
    anomaly_indices = np.where(anomaly_flags)[0]
    report = []
    for idx in anomaly_indices:
        anomaly_info = {
            'index': idx,
            'error': test_mse[idx],
            'timestamp': api_logs.iloc[idx + window_size - 1]['Timestamp'] if 'Timestamp' in api_logs.columns else None,
            'context': api_logs.iloc[idx:idx + window_size].to_dict(orient='records')
        }
        report.append(anomaly_info)
    return report


if __name__ == "__main__":
    # Parameters
    file_path = 'synthetic_logs_dataset.csv'  # Update this path as necessary
    window_size = 20
    test_size = 0.2
    val_size = 0.1
    threshold_percentile = 95
    
    # Stage 1: Load and preprocess data
    data_scaled, api_logs = load_and_preprocess_data(file_path)
    
    # Stage 2: Create sequences from the scaled data
    sequences = create_sequences(data_scaled, window_size)
    print("Shape of sequences:", sequences.shape)
    
    # Split data into training, validation, and test sets (without shuffling for time series)
    X_train, X_test = train_test_split(sequences, test_size=test_size, shuffle=False)
    X_train, X_val = train_test_split(X_train, test_size=val_size, shuffle=False)
    print("Training shape:", X_train.shape, "Validation shape:", X_val.shape, "Testing shape:", X_test.shape)
    
    # Stage 3: Build the LSTM autoencoder model with attention
    model = build_lstm_autoencoder((window_size, data_scaled.shape[1]))
    model.summary()
    
    # Stage 4: Train the model
    history = train_model(model, X_train, X_val)
    
    # Plot training history
    plt.figure(figsize=(8, 4))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel("Epoch")
    plt.ylabel("Loss (MSE)")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.show()
    
    # Stage 5: Evaluate the model on the test set
    test_mse, threshold, anomaly_flags = evaluate_model(model, X_test, threshold_percentile)
    print(f"Anomaly detection threshold ({threshold_percentile}th percentile): {threshold:.4f}")
    print(f"Number of anomalies detected: {sum(anomaly_flags)} out of {len(test_mse)}")
    
    # Stage 6: Generate a detailed anomaly report
    anomaly_report = generate_anomaly_report(test_mse, threshold, anomaly_flags, api_logs, window_size)
    
    # Save the model using the Keras recommended format
    model.save('anomaly_detection_model.keras')
    
    # Optionally, visualize a few anomalies
    plt.figure(figsize=(12, 6))
    anomaly_indices = np.where(anomaly_flags)[0]
    for i in range(min(3, len(anomaly_indices))):
        idx = anomaly_indices[i]
        plt.clf()
        # Predict reconstruction for a single anomaly sequence
        reconstructed = model.predict(X_test[np.newaxis, idx, :, :])
        plt.plot(X_test[idx].flatten(), label='Original')
        plt.plot(reconstructed.flatten(), label='Reconstructed')
        plt.title(f"Anomaly {i+1} - Reconstruction Error: {test_mse[idx]:.4f}")
        plt.legend()
        plt.show()
    
    # Print the detailed anomaly report
    print("\nDetailed Anomaly Report:")
    for item in anomaly_report:
        print("\n------------------------------")
        print(f"Anomaly Index: {item['index']}")
        print(f"Timestamp: {item['timestamp']}")
        print(f"Reconstruction Error: {item['error']:.4f}")
        print("Contextual Data:")
        for context in item['context']:
            print(context)
```
